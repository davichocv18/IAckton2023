```python
import pandas as pd
from IPython.display import display

# Assuming "releases" is the name of the column containing the nested dictionaries
new_df = df.explode("releases")

# Create a DataFrame for the "tender" column
selected_Tender = ["tender"]
table_Tender = pd.DataFrame.from_records(new_df["releases"].to_list(), columns=selected_Tender)

# Get the values from the "procuringEntity," "tenderers," and "value" columns
table_Tender['procuringEntity'] = table_Tender['tender'].apply(lambda x: x['procuringEntity'])
table_Tender['tenderers'] = table_Tender['tender'].apply(lambda x: x['tenderers'])
table_Tender['value'] = table_Tender['tender'].apply(lambda x: x['value'])

# Display the resulting DataFrame with the selected columns using IPython.display
display(table_Tender[['procuringEntity', 'tenderers', 'value']])
```

**Realizaremos un extracion de *Contracts* para poder clasificar el monto del contrato, para tambien vizualizar si este tambien se encuentra finalizado o en proceso.** Y lo podres vizualizar el saber cuantos contratos terminaron a travez de una tabla.

```python
import pandas as pd
from IPython.display import display

# Crear un DataFrame con la columna "contracts"
table_Contract = pd.DataFrame(new_df["releases"].to_list())

# Obtener los valores de la columna "id", "status" y "value" de los contratos
table_Contract['contract_id'] = table_Contract['contracts'].apply(lambda x: x[0]['id'] if isinstance(x, list) and len(x) > 0 else "En proceso")
table_Contract['contract_status'] = table_Contract['contracts'].apply(lambda x: x[0]['status'] if isinstance(x, list) and len(x) > 0 else "Sin terminar")
table_Contract['contract_value'] = table_Contract['contracts'].apply(lambda x: x[0]['value'] if isinstance(x, list) and len(x) > 0 else "Sin definir")

# Función para determinar si un contrato está terminado (1) o no (0)
def is_contract_terminated(status):
    return 1 if status == "terminated" else 0

# Aplicar la función para crear la columna "contract_terminated"
table_Contract['contract_terminated'] = table_Contract['contract_status'].apply(is_contract_terminated)

# Reemplazar los valores donde no hay datos o contrato sin terminar con 0
table_Contract['contract_id'] = table_Contract.apply(lambda row: '0' if row['contract_id'] == "En proceso" else row['contract_id'], axis=1)
table_Contract['contract_value'] = table_Contract.apply(lambda row: 0 if row['contract_status'] == "Sin terminar" else row['contract_value'], axis=1)

# Mostrar los datos en una tabla
display(table_Contract[['contract_id', 'contract_status', 'contract_value', 'contract_terminated']])




```

Primer Grafo Tender:  Usamos los datos de Nombre de Procaridad contratante, Nombre del Provedor, y Valor del Contrato para su relación.

```python
import pandas as pd
import networkx as nx
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import dash
from dash import dcc, html
from dash import dash_table
import plotly.express as px

# Crear un grafo vacío
G1 = nx.Graph()

# Iterar a través de las filas de tu DataFrame
for index, row in table_Tender.iterrows():
    procuring_entity = row['procuringEntity']['name']
    tenderers = [t['name'] for t in row['tenderers']]
    value = float(row['value']['amount'])

    # Agregar nodos para el procuringEntity y los tenderers con etiquetas y valores
    G1.add_node(procuring_entity, node_type='procuringEntity', label=procuring_entity, value=value)
    for tenderer in tenderers:
        G1.add_node(tenderer, node_type='tenderer', label=tenderer, value=value)

    # Agregar una arista entre el procuringEntity y cada tenderer
    for tenderer in tenderers:
        G1.add_edge(procuring_entity, tenderer, value=value)

# Crear un grafo dirigido para Plotly
pos = nx.spring_layout(G1, k=0.15)
edge_x = []
edge_y = []

for edge in G1.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines',
    name='Relaciones'  # Nombre personalizado para las trazas de enlaces
)

node_x = []
node_y = []
node_colors = []  # Lista para los colores de los nodos
node_values = []  # Lista para los valores de los nodos

for node in G1.nodes():
    x, y = pos[node]
    node_x.append(x)
    node_y.append(y)
    node_values.append(G1.nodes[node]['value'])
    if G1.nodes[node]['node_type'] == 'procuringEntity':
        node_colors.append('purple')  # Color morado para procuring_entity
    else:
        node_colors.append('orange')  # Color naranja para tenderers

# Filtrar nodos por tipo
procuring_entity_nodes = [node for node in G1.nodes() if G1.nodes[node]['node_type'] == 'procuringEntity']
tenderer_nodes = [node for node in G1.nodes() if G1.nodes[node]['node_type'] == 'tenderer']

# Crear trazas separadas para compradores y proveedores
procuring_entity_trace = go.Scatter(
    x=[node_x[i] for i, node in enumerate(G1.nodes()) if node in procuring_entity_nodes],
    y=[node_y[i] for i, node in enumerate(G1.nodes()) if node in procuring_entity_nodes],
    mode='markers',
    hoverinfo='text',
    text=[f'{G1.nodes[node]["label"]}<br># of connections: {len(list(G1.neighbors(node)))}<br>Value:$ {G1.nodes[node]["value"]}' for node in procuring_entity_nodes],
    marker=dict(
        color='purple',
        size=10,
    ),
    name='Entidades Contratantes'  # Nombre para la leyenda
)

tenderer_trace = go.Scatter(
    x=[node_x[i] for i, node in enumerate(G1.nodes()) if node in tenderer_nodes],
    y=[node_y[i] for i, node in enumerate(G1.nodes()) if node in tenderer_nodes],
    mode='markers',
    hoverinfo='text',
    text=[f'{G1.nodes[node]["label"]}<br># of connections: {len(list(G1.neighbors(node)))}<br>Value:$ {G1.nodes[node]["value"]}' for node in tenderer_nodes],
    marker=dict(
        color='orange',
        size=10,
    ),
    name='Proveedores'  # Nombre para la leyenda
)

# Crear una lista de diccionarios con la información de los nodos
node_info = []

for node in G1.nodes():
    info = {
        'Label': G1.nodes[node]['label'],
        'Number of Connections': len(list(G1.neighbors(node))),
        'Amount Value': G1.nodes[node]['value']
    }
    node_info.append(info)

# Crear la figura de red usando Plotly
fig_network1 = go.Figure(data=[edge_trace, procuring_entity_trace, tenderer_trace],  # Agregar las dos trazas
                       layout=go.Layout(
                           showlegend=True,
                           hovermode='closest',
                           margin=dict(b=0, l=0, r=0, t=0),
                           xaxis=dict(showticklabels=False),  # Eliminar etiquetas en el eje x
                           yaxis=dict(showticklabels=False)   # Eliminar etiquetas en el eje y
                       ))

# Configurar la leyenda con información adicional
fig_network1.update_layout(
    legend=dict(
        title='Lista Interactiva',  # Título de la leyenda
        traceorder='normal',           # Orden en que se muestran las trazas (normal o reversed)
        font=dict(size=12),            # Tamaño de fuente de la leyenda
    ),
)

# Agregar una anotación para el texto personalizado en la leyenda
fig_network1.add_annotation(
    text="Relación Licitaciones 2023",
    x=0,  # Ajusta la posición horizontal del texto en la leyenda
    y=1.1,  # Ajusta la posición vertical del texto en la leyenda
    showarrow=False,  # No mostrar una flecha de anotación
    font=dict(size=24)  # Ajusta el tamaño de fuente del texto en la leyenda
)

# Mostrar la figura
fig_network1.show()

# Calcular el valor total para cada comprador
procuring_entity_values = {}  # Un diccionario para almacenar los valores totales de las entidades contratantes

for node in G1.nodes():
    if G1.nodes[node]['node_type'] == 'procuringEntity':
        # Inicializar el valor total en 0 para cada entidad contratante
        procuring_entity_values[node] = 0

# Calcular el valor total sumando los valores de los nodos conectados a cada entidad contratante
for node in G1.nodes():
    if G1.nodes[node]['node_type'] == 'tenderer':
        neighbors = list(G1.neighbors(node))
        for neighbor in neighbors:
            if G1.nodes[neighbor]['node_type'] == 'procuringEntity':
                procuring_entity_values[neighbor] += G1.nodes[node]['value']

# Crear un DataFrame con los valores totales
total_values_df = pd.DataFrame.from_dict(procuring_entity_values, orient='index', columns=['Valor Total Entidad Contratante'])
total_values_df.index.name = 'Entidad Contratante'  # Definir un nombre para el índice

# Ordenar la tabla de mayor a menor valor total
total_values_df = total_values_df.sort_values(by='Valor Total Entidad Contratante', ascending=False)

# Agregar el símbolo "$" antes del valor
total_values_df['Valor Total Entidad Contratante'] = total_values_df['Valor Total Entidad Contratante'].apply(lambda x: f"${x:,.2f}")

# Crear una aplicación Dash
app = dash.Dash(__name__)

# Definir el diseño de la aplicación
app.layout = html.Div([
    html.H1("Tabla de Valores Totales de Entidades Contratantes"),
    dash_table.DataTable(
        id='table',
        columns=[
            {'name': 'Entidad Contratante', 'id': 'Entidad Contratante'},
            {'name': 'Valor Total Entidad Contratante', 'id': 'Valor Total Entidad Contratante'},
        ],
        data=total_values_df.reset_index().to_dict('records'),
        sort_action='native',  # Permite la ordenación nativa de Dash
    ),
])

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)

# Guardar el gráfico en un archivo GEXF para su visualización en herramientas como Gephi
nx.write_gexf(G1, "red_de_contrataciones_Tender.gexf")
```

```python
import json
import pandas as pd
from IPython.display import display

# Ruta al archivo JSON
archivo_json = '/content/releases_2023_obra_artistica_cientifica_o_literaria.json'

# Cargar el archivo JSON
with open(archivo_json, 'r', encoding='utf-8') as archivo:
    data = json.load(archivo)

# Convertir el JSON en un DataFrame de Pandas
df = pd.DataFrame(data)

# Seleccionar solo la columna "releases"
releases = df['releases']

# Assuming "releases" is the name of the column containing the nested dictionaries
new_df = df.explode("releases")

# Crear una lista de diccionarios para los datos de awards y buyers
combined_data = []
for release in new_df["releases"]:
    awards = release.get("awards", [])
    buyer = release.get("buyer", {})
    contracts = release.get("contracts", [])

    for award in awards:
        award_data = {
            "ID": award["id"],
            "Date": award["date"],
            "Amount": award["value"]["amount"],
            "Supplier Name": award["suppliers"][0]["name"],
            "Description": award["description"],
            "Buyer ID": buyer.get("id"),
            "Buyer Name": buyer.get("name"),
            "Contracts": contracts[0]["id"] if contracts else 0  # Use 0 if "contracts" is empty
        }
        combined_data.append(award_data)

# Crear un DataFrame para los datos combinados de awards y buyers
combined_df = pd.DataFrame(combined_data)

# Filtrar las filas donde "Contracts" no sea igual a 0
combined_df = combined_df[combined_df["Contracts"] != 0]

# Display the resulting table with headers using IPython.display
display(combined_df[['ID', 'Date', 'Amount', 'Supplier Name', 'Description', 'Buyer ID', 'Buyer Name', 'Contracts']])


```

**GRAFO DE AWARDS**

```python
import json
import pandas as pd
import networkx as nx
import plotly.graph_objects as go
from collections import defaultdict

# Ruta al archivo JSON
archivo_json = '/content/releases_2023_obra_artistica_cientifica_o_literaria.json'

# Cargar el archivo JSON
with open(archivo_json, 'r', encoding='utf-8') as archivo:
    data = json.load(archivo)

# Convertir el JSON en un DataFrame de Pandas
df = pd.DataFrame(data)

# Seleccionar solo la columna "releases"
releases = df['releases']

# Assuming "releases" is the name of the column containing the nested dictionaries
new_df = df.explode("releases")

# Crear una lista de diccionarios para los datos de awards y buyers
combined_data = []
for release in new_df["releases"]:
    awards = release.get("awards", [])
    buyer = release.get("buyer", {})
    contracts = release.get("contracts", [])

    for award in awards:
        award_data = {
            "Buyer ID": buyer.get("name"),
            "Supplier Name": award["suppliers"][0]["name"],
            "Amount": award["value"]["amount"]
        }
        combined_data.append(award_data)

# Crear un DataFrame para los datos combinados de awards y buyers
combined_df = pd.DataFrame(combined_data)

# Filtrar las filas donde "Amount" no sea igual a 0
combined_df = combined_df[combined_df["Amount"] != 0]

# Calcular el valor total de enlaces para cada entidad compradora
buyer_total_amounts = defaultdict(int)
for _, row in combined_df.iterrows():
    buyer_id = row["Buyer ID"]
    amount = row["Amount"]
    buyer_total_amounts[buyer_id] += amount

# Crear un grafo dirigido con los datos proporcionados
G2 = nx.DiGraph()
for _, row in combined_df.iterrows():
    buyer_id = row["Buyer ID"]
    supplier_name = row["Supplier Name"]
    amount = row["Amount"]
    G2.add_node(buyer_id, type='buyer', label=buyer_id, value=buyer_total_amounts[buyer_id])
    G2.add_node(supplier_name, type='supplier', label=supplier_name, value=amount)
    G2.add_edge(buyer_id, supplier_name)

# Crear dos listas separadas para compradores y proveedores
buyer_nodes = [node for node in G2.nodes() if G2.nodes[node]['type'] == 'buyer']
supplier_nodes = [node for node in G2.nodes() if G2.nodes[node]['type'] == 'supplier']

# Crear trazas para compradores y proveedores
buyer_trace = go.Scatter(
    x=[pos[node][0] for node in buyer_nodes],
    y=[pos[node][1] for node in buyer_nodes],
    mode='markers',
    hoverinfo='text',
    marker=dict(
        color='purple',
        size=10,
    ),
    text=[f'Buyer: {node}<br>Total Amount: {buyer_total_amounts[node]}' for node in buyer_nodes],
    name='Compradores'  # Nombre de la traza de compradores
)

supplier_trace = go.Scatter(
    x=[pos[node][0] for node in supplier_nodes],
    y=[pos[node][1] for node in supplier_nodes],
    mode='markers',
    hoverinfo='text',
    marker=dict(
        color='orange',
        size=10,
    ),
    text=[f'Supplier: {node}<br>Amount: {G2.nodes[node]["value"]}' for node in supplier_nodes],
    name='Proveedores'  # Nombre de la traza de proveedores
)

# Crear la figura de red usando Plotly con las dos trazas
fig_network2 = go.Figure(data=[edge_trace, buyer_trace, supplier_trace],
                       layout=go.Layout(
                           showlegend=True,
                           hovermode='closest',
                           margin=dict(b=0, l=0, r=0, t=0),
                           xaxis=dict(showticklabels=False),
                           yaxis=dict(showticklabels=False)
                       ))

# Agregar una anotación para el texto personalizado en la leyenda
fig_network2.add_annotation(
    text="Relación AWARDS 2023",
    x=0,
    y=1.1,
    showarrow=False,
    font=dict(size=24)
)

# Mostrar el gráfico interactivo
fig_network2.show()

# Guardar el gráfico en un archivo GEXF para su visualización en herramientas como Gephi
nx.write_gexf(G2, "red_de_contrataciones.gexf")




```
